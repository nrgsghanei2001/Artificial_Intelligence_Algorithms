```python
import random
import math
import numpy as np
```

calculate fitness which gives the number of roles should be used according to given state

```python
def evaluation_function(state, stock_len):
    num_rolls, sum = 1, 0
    roll = []
    for i in range(len(state)):
        sum += state[i]
        if sum > stock_len:         # if cuts in a roll is more than capacity of roll increase number of rolls
            roll = []                          # free the roll
            roll.append(state[i])         # add this stock to roll
            sum = state[i]
            num_rolls += 1
        else:
            roll.append(state[i])
    
    return num_rolls
```

Find neighbor of a given state 

```python
def find_neighbor(state):      # choose 3 pair of random index and swap their elements in state array
    len_cuts = len(state)
    i = random.randint(0, len_cuts-1)
    j = random.randint(0, len_cuts-1)
    i1 = random.randint(0, len_cuts-1)
    j1 = random.randint(0, len_cuts-1)
    i2 = random.randint(0, len_cuts-1)
    j2 = random.randint(0, len_cuts-1)

    temp = state[i]
    state[i] = state[j]
    state[j] = temp

    temp = state[i1]
    state[i1] = state[j1]
    state[j1] = temp

    temp = state[i2]
    state[i2] = state[j2]
    state[j2] = temp

    return state
    
```

generate an initial state which is not random and is closer to result of problem

```python
# Define a function to generate a better initial state
def generate_better_initial_state(requests, roll_length):

  # Sort the requests in descending order
  requests = sorted(requests, reverse=True)

  # Initialize an empty list of rolls
  rolls = [[roll_length]]

  # For each request in the sorted list
  for request in requests:
    # Find the first roll in the list that can accommodate the request
    found = False
    for roll in rolls:
      if roll[0] >= request:
        found = True
        break

    # If such a roll exists, add the request to that roll and update the remaining space
    if found:
      roll.append(request)
      roll[0] -= request

    # If no such roll exists, create a new roll with the request and add it to the list
    else:
      roll = [roll_length - request, request]
      rolls.append(roll)
  state = []
  for roll in rolls:
    for i in range(1, len(roll)):
      state.append(roll[i])

  # Return the vector of counts as the initial state
  return state
```

Hill Climbing Algorithm

main algorithm

```python
def hill_climbing(cuts, stock_len, iterations):
    current_state = cuts.copy()
    best_result = []
    random.shuffle(current_state)       # generate a random initial state
    current_value = evaluation_function(current_state, stock_len)   # find the value of current state

    end = False      # stop condition
    num_iteration_withput_exchange = 0       # number of iteration that the state does not exchange with its neighbor
    
    while not end:
        neighbor = find_neighbor(current_state)                 # find the neighbor of current state and its fitness
        neighbor_value = evaluation_function(neighbor, stock_len)

        if neighbor_value < current_value:   # if number of used rolls of neighbor is better than current state, exchange them
            current_state = neighbor
            best_result = neighbor[:]
            current_value = neighbor_value
            num_iteration_withput_exchange = 0  # exchange done
        
        else:
            num_iteration_withput_exchange += 1
            if num_iteration_withput_exchange > iterations :      # if limit of iterations reached end the loop and show founded result
                end = True

    return best_result    # the best result which algorithm found 
```

</em>Simulated Annealing Algorithm</em>

A function which gives the acceptance probability of a new state 

```python
def acceptance_probability(current_value, new_value, temprature):
    delta = (new_value - current_value)/current_value

    if delta < 0:           # if the value of new state is better than current state, exchange them
        return 1

    r = np.random.rand()     # else: generate a random number between 0, 1
    p = np.exp(-delta/temprature)  # calculate acceptance probability
    
    if (p > r):        # accept new state with calculated probability 
        return 2 
    return 0
```

A function that reducing the temprature over time

```python
def cooling_function(current_temp):
    alpha = 0.99
    new_temp = current_temp * alpha
    return new_temp
```

```python
def simulated_annealing(cuts, stock_len, temprature, TL, iteration):
    current_state = cuts.copy()
    random.shuffle(current_state)          # generate a random initial state
    current_value = evaluation_function(current_state, stock_len)           # find the value of current state
    best_result = current_state[:]         # save the best result 
    best_cost = math.inf     # save the best cost
    num_iteration = 0
    while temprature > TL or num_iteration < iteration:    # stop condition: when tempreture reachs TL by cooling function
        num_iteration += 1
        neighbor = find_neighbor(current_state)
        neighbor_value = evaluation_function(neighbor, stock_len)

        p = acceptance_probability(current_value, neighbor_value, temprature)  # the result clarifies that we should move or not
        if p == 1:   # p=1 : exchange with new state, new state has better fitness
            current_state = neighbor
            current_value = neighbor_value
            best_result = neighbor[:]
            best_cost =  current_value

        elif p == 2:     # exchange with new state, new state is worst than current state
            current_state = neighbor
        # p=0, don't exchange states

        temprature = cooling_function(temprature)   # lower the temprature in each iteration

    return best_result
```

Genetic Algorithm

Create random population

```python
def create_population(population_len, cuts):
    population = []
    state = cuts
    for i in range(population_len):
        state = sorted(state, key=lambda x: random.random())   # shuffle the cuts list and create a random gen
        population.append(state)

    return population
```

select pair of parents

```python
def parent_selection(population, stock_len, selection_rate):
    # choose random n parent for each parent
    father_candidates = []
    for i in range(selection_rate):
        state = random.choice(population)
        father_candidates.append(state)

    mother_candidates = []
    for i in range(selection_rate):
        state = random.choice(population)
        mother_candidates.append(state)

    # choose the best one of n candidates to be a parent
    father_val = math.inf
    father_ind = 0
    for i in range(selection_rate):
        val = evaluation_function(father_candidates[i], stock_len)
        if val < father_val:
            father_val = val
            father_ind = i

    mother_val = math.inf
    mother_ind = 0
    for i in range(selection_rate):
        val = evaluation_function(mother_candidates[i], stock_len)
        if val < mother_val:
            mother_val = val
            mother_ind = i

    return father_candidates[father_ind], mother_candidates[mother_ind]
```

Cross over parent and generate 2 children

```python
def cross_over(father, mother, cuts, occurrence):
    len_cuts = len(cuts)
    set_cut = set(cuts)       # find keys of cuts(unique cuts)
    item_count = {item:0 for item in set_cut}       # define a dictionary for mapping each cut length with its count

    child1, child2 = [], []
    random_point = random.randint(0, len_cuts-1)    # select a random point and cut the father from there

    for i in range(0, random_point):
            child1.append(father[i])
            item_count[father[i]] += 1       # increase the counter of used cut length

    for i in range(random_point, len_cuts):         # add the rest of mother from random point if its not used before from father
        if item_count[mother[i]] < occurrence[mother[i]]:   # if the cut length does not occured the same as in initial cuts
            child1.append(mother[i])
            item_count[mother[i]] += 1

    for i in range(0, random_point):     # add mother from first to the random point
        if item_count[mother[i]] < occurrence[mother[i]]:
            child1.append(mother[i])
            item_count[mother[i]] += 1

    item_count = {item:0 for item in set_cut}        # do the same but this time cross mother first for the child2
    for i in range(0, random_point):
            child2.append(mother[i])
            item_count[mother[i]] += 1

    for i in range(random_point, len_cuts):
        if item_count[father[i]] < occurrence[father[i]]:
            child2.append(father[i])
            item_count[father[i]] += 1

    for i in range(0, random_point):
        if item_count[father[i]] < occurrence[father[i]]:
            child2.append(father[i])
            item_count[father[i]] += 1

    return child1, child2
```

Mutation Function

```python
def mutation(child, cuts):
    len_cuts = len(cuts)
    i = random.randint(0, len_cuts-1)
    j = random.randint(0, len_cuts-1)

    # choose 2 random points and replace them with each other
    temp = child[i]
    child[i] = child[j]
    child[j] = temp
   
    return child
```

```python
def genetic(cuts, occurrence, stock_len, population_size, crossover_rate, mutation_rate, replacement_rate, selection_rate, generation_limit):
    best_cost = math.inf
    best_result = []

    population = create_population(population_size, cuts)     # create a population of gens with size of population size
    num_generation = 0
    while num_generation < generation_limit:   # stop condition
        num_generation += 1          # increase number of generated generations
        children_population = []          # save new population of children
        for i in range(population_size//2):        # create a children population the same size as parents population
            
            father , mother = parent_selection(population, stock_len, selection_rate)       # select 2 parents from population

            p = random.random()
            if p > crossover_rate:
                child1, child2 = cross_over(father, mother, cuts, occurrence)  # cross over parents and generate 2 children
            else:     # children are the same as parents
                child1 = father
                child2 = mother

            p = random.random()        # mutate each child with a probability of mutation rate
            if p > mutation_rate:
                child1 = mutation(child1, cuts)
            p = random.random()
            if p > mutation_rate:
                child2 = mutation(child2, cuts)

            children_population.append(child1)      # add children to new population
            children_population.append(child2)

        parent_values = []
        child_values = []
        for i in range(len(population)):             # save all values of parents and children in 2 value lists
            x = evaluation_function(population[i], stock_len)
            y = evaluation_function(children_population[i], stock_len)
            parent_values.append(x)
            child_values.append(y)
        
        num_replace = replacement_rate * len(population)    # find the number of parents that should be replaced with children in population

        for i in range(int(num_replace)):        # replace as the size of replacements in population
            max_parent_ind = parent_values.index(max(parent_values))  # find the worst parent
            min_child_ind = child_values.index(min(child_values))      # find best children

            population.pop(max_parent_ind)      # remove worst parent from population
            population.append(children_population[min_child_ind])  # add best children to population

            children_population.pop(min_child_ind)  # remove best children from population of children
            parent_values.pop(max_parent_ind)   # remove worst parent value from parents values
            child_values.pop(min_child_ind)  # remove best children value from children values

    # find the best result in population
    for i in range(len(population)):
        if evaluation_function(population[i], stock_len) < best_cost:
            best_cost = evaluation_function(population[i], stock_len)
            best_result = population[i]

    return best_result
```
